```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn for enhanced visualization

df = pd.read_csv('Titanic-Dataset.csv')
df.head()
df.head(100)

df.shape
df.info()
df.isnull().sum()
df.describe()

num_only = df.select_dtypes(include=[np.number])
corr = num_only.corr(numeric_only=True)

plt.figure(figsize=(8, 6)) # Increased figure size for better readability
sns.heatmap(corr, cmap='coolwarm', vmin=-1, vmax=1, annot=True, fmt=".2f") # Use seaborn.heatmap with annotation
plt.title('Heatmap Korelasi (numerik)')
plt.tight_layout()
plt.show()

display(corr['Survived'].sort_values(ascending=False))

num_cols = ['Age','Fare']
for col in num_cols:
    plt.figure()
    plt.hist(df[col], bins=30)
    plt.title(f'Distribusi {col}')
    plt.xlabel(col); plt.ylabel('Frekuensi')
    plt.show()

plt.figure()
plt.boxplot(df['Fare'].dropna(), vert=True)
plt.title('Boxplot Fare')
plt.ylabel('Fare')
plt.show()

cat_cols = ['Survived','Pclass','Sex','Embarked']

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    df[col].value_counts(dropna=False).plot(kind='bar', ax=axes[i])
    axes[i].set_title(f'Distribusi {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Jumlah')

plt.tight_layout()
plt.show()

plt.figure()
surv_map = {0:'No', 1:'Yes'}
plt.scatter(df['Age'], df['Fare'], alpha=0.7)
plt.xlabel('Age'); plt.ylabel('Fare')
plt.title('Scatter Age vs Fare')
plt.show()
```

## PRAKTIKUM 1 (JS3)

```python
import pandas as pd
from google.colab import files
import io
import matplotlib.pyplot as plt
import seaborn as sns

# Upload the dataset
uploaded = files.upload()

# Assuming you uploaded 'Titanic-Dataset (1).csv'
# If you uploaded a different file, change 'Titanic-Dataset (1).csv' to your filename
file_name = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

# Display the first few rows
display(df.head())
df.info()
df.isnull().sum()

# Age - mean
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin - "DECK"
df['Cabin'].fillna(value="DECK", inplace=True)

# Embarked - modus
df['Embarked'].fillna(value=df['Embarked'].mode, inplace=True)

# Display the first few rows
display(df.head())
df.info()
df.isnull().sum()

# === EKSTRAKSI FITUR ===

# 1. FamilySize: jumlah keluarga di kapal (SibSp + Parch + diri sendiri)
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

# 2. Title: ambil gelar dari kolom Name (Mr, Mrs, Miss, Master, dll.)
df["Title"] = df["Name"].str.extract(r',\s*([^\.]+)\.', expand=False)
title_mapping = {
    "Mlle": "Miss", "Ms": "Miss", "Mme": "Mrs",
    "Lady": "Rare", "Countess": "Rare", "Capt": "Rare", "Col": "Rare",
    "Don": "Rare", "Dr": "Rare", "Major": "Rare", "Rev": "Rare",
    "Sir": "Rare", "Jonkheer": "Rare", "Dona": "Rare"
}
df["Title"] = df["Title"].replace(title_mapping)

# 3. AgeBin: kelompokkan usia menjadi kategori
df["AgeBin"] = pd.cut(df["Age"], bins=[0, 12, 18, 35, 60, 100],
                      labels=["Child", "Teen", "YoungAdult", "Adult", "Senior"])

# 4. CabinDeck: ambil huruf pertama dari kolom Cabin
df["CabinDeck"] = df["Cabin"].astype(str).str[0]

# 5. Fare per person: bagi harga tiket dengan jumlah keluarga
df["FarePerPerson"] = df["Fare"] / df["FamilySize"]

print("✅ Hasil Ekstraksi fitur ")
df[["Name", "Title", "Age", "AgeBin", "FamilySize", "Cabin", "CabinDeck", "FarePerPerson"]].head()

# === VISUALISASI FITUR HASIL EKSTRAKSI ===

plt.figure(figsize=(12, 6))
sns.barplot(x="FamilySize", y="Survived", data=df)
plt.title("Survival Rate berdasarkan Family Size")
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x="Title", y="Survived", data=df,
            order=df.groupby("Title")["Survived"].mean().sort_values().index)
plt.title("Survival Rate berdasarkan Title")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x="AgeBin", y="Survived", data=df,
            order=["Child", "Teen", "YoungAdult", "Adult", "Senior"])
plt.title("Survival Rate berdasarkan Kelompok Usia (AgeBin)")
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x="CabinDeck", y="Survived", data=df,
            order=df.groupby("CabinDeck")["Survived"].mean().sort_values().index)
plt.title("Survival Rate berdasarkan Cabin Deck")
plt.show()
```

## TUGAS PRAKTIKUM (JS3)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import numpy as np

# Load the dataset
df_wbc = pd.read_csv('/content/wbc.csv')

# 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.
# Identify and drop the irrelevant columns ('id' and 'Unnamed: 32')
df_wbc = df_wbc.drop(['id', 'Unnamed: 32'], axis=1)
print("✅ Kolom 'id' dan 'Unnamed: 32' telah dihapus.")

# 2. Lakukan proses encoding pada kolom "diagnosis".
le = LabelEncoder()
df_wbc['diagnosis'] = le.fit_transform(df_wbc['diagnosis'])
print("✅ Kolom 'diagnosis' telah di-encoding.")

# 3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.
std = StandardScaler()
# Select only numeric columns, excluding the target variable 'diagnosis'
numeric_cols = df_wbc.select_dtypes(include=np.number).columns.tolist()
numeric_cols.remove('diagnosis')
df_wbc[numeric_cols] = std.fit_transform(df_wbc[numeric_cols])
print("✅ Kolom numerik telah distandarisasi.")

# 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.
# Define features (X) and target (y)
X = df_wbc.drop('diagnosis', axis=1)
y = df_wbc['diagnosis']

# Perform stratified split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
print("✅ Data telah dibagi menjadi data latih (80%) dan data uji (20%) secara stratified.")

print("\n=== Informasi Data Setelah Preprocessing dan Splitting ===")
print(f"Ukuran data asli: {df_wbc.shape}")
print(f"Ukuran data latih (X_train): {X_train.shape}, Ukuran label latih (y_train): {y_train.shape}")
print(f"Ukuran data uji (X_test): {X_test.shape}, Ukuran label uji (y_test): {y_test.shape}")

print("\n=== Contoh Data Latih (X_train) ===")
display(X_train.head())
print("\n=== Contoh Label Latih (y_train) ===")
display(y_train.head())
```
